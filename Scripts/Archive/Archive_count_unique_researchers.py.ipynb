```python
import Functions.StringComp as SC
import Models.ImportModel as IM
import pandas as pd
import json
import re

VR = IM.original()
```

```python
df = VR.df[['Dnr', 'Efternamn', 'Förnamn']]
# rename columns
df.columns = ['app_id', 'surname', 'name']
```

```python
try:
    # import the latest duplicate authors file
    duplicate_authors = pd.read_excel('duplicates_backup.xlsx')
    last_row = duplicate_authors.iloc[-1]
    last_app_ids = re.sub(r"[\[\]\']", "", last_row['app_ids']).split(", ")
    last_app_id = last_app_ids[0]

    # in case we aborted in the middle of an interation, we remove the last row and redo it
    duplicate_authors = duplicate_authors.iloc[:-1 , :]
    start_app_id = last_app_id
except:
    # create the new dataframe
    duplicate_authors = pd.DataFrame(columns=['name', 'surname', 'app_ids'])
    start_app_id = None

    if start_app_id == None:
        start_app_id = df['app_id'].iloc[0]

print("start_id is", start_app_id)


limit = None
if limit == None:
    limit = len(df)

discovered_duplicates = []
for tup in duplicate_authors.itertuples():
    list = re.sub(r"[\[\]\']", "", tup[3]).split(", ")
    for a in list:
        if a in discovered_duplicates:
            raise ValueError("DUPLICATE APPID IN THE WRONG SENSE")
        else:
            discovered_duplicates.append(a)

checking = True

for rowtup in df.itertuples():
   if rowtup[0] >= limit:
       break
   o = {}
   o['app_id'], o['surname'], o['name'] = rowtup[1], rowtup[2], rowtup[3]

   if start_app_id == o['app_id'] and checking == True:
       checking = False
   #print("Name is", o['name'])
   if checking == True:
       continue

   dup = False
   app_ids = [o['app_id']]

   for rowtup_inner in df.itertuples():
       i = {}
       i['app_id'], i['surname'], i['name'] =  rowtup_inner[1], rowtup_inner[2], rowtup_inner[3]

       if i['app_id'] in discovered_duplicates:
           continue

       if o['app_id'] != i['app_id']:

           if o['name'] == i['name'] and o['surname'] == i['surname']:
               if dup == False:
                   print(o['app_id'], o['name'], o['surname'])
                   discovered_duplicates.append(o['app_id'])
                   dup = True
               print(i['app_id'], i['name'], i['surname'])
               app_ids.append(i['app_id'])
               discovered_duplicates.append(i['app_id'])

   if len(app_ids) > 1:
       temporary_df = pd.DataFrame([[o['name'], o['surname'], app_ids]],
                                   columns=['name', 'surname', 'app_ids'])
       duplicate_authors = duplicate_authors.append(temporary_df, ignore_index=True)
```

```python
#print(duplicate_authors)

# Export the list of duplicate authors

# Create a new excel workbook
writer = pd.ExcelWriter('duplicates_2.xlsx', engine='xlsxwriter')
duplicate_authors.to_excel(writer, sheet_name="duplicates", index=False)

writer.save()
```
